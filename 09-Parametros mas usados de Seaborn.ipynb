```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tip = sns.load_dataset('tips')
tip.head(5)
```

```python
sns.displot(data=tip,x='total_bill')
plt.show()
```

```python
sns.displot(data=tip,x='total_bill',y='tip')
plt.show()
```

```python
sns.displot(data=tip,x='total_bill',y='tip',hue='sex')
plt.show()
```

```python
sns.displot(data=tip,x='total_bill',hue='sex',kind='kde')
plt.show()
```

```python
sns.displot(data=tip,x='total_bill',hue='sex',
            kind='kde',legend=True,palette='dark',
            alpha = 0.5)
plt.show()
```
