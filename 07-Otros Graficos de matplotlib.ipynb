```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

## **Histograma** 
Muestra la distribucion de los datos

```python
plt.hist(data,bins=10,histtype='step')
plt.show()
```

```python
plt.hist(data,bins=10,histtype='bar')
plt.show()
```

## **Boxplot**
Muestra la distribucion de mis datos en Quartiles

```python
data2 = np.append(data,200)
plt.boxplot(data2,vert=False,patch_artist=True,notch=True,showfliers=True)
plt.show()
```

```python
plt.boxplot(data2,vert=False,patch_artist=True,notch=False,showfliers=False)
plt.show()
```

## **Sactter plot** 
Muestra la relacion entre dos variables

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y,s=area,c=colors,marker='.',alpha=0.5)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```
