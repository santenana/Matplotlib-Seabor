```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tip = sns.load_dataset('tips')
tip.head(5)
```

```python
sns.countplot(data=tip, x='day',hue='sex')
plt.show()
```

```python
sns.countplot(data=tip, y='day',hue='sex')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tip, x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tip, x='day',y='total_bill',hue='sex',dodge=True,marker='*')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tip, x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tip, x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
sns.catplot(data=tip, x='day',y='total_bill',hue='sex',
            dodge=True,kind='strip',col='time')
plt.show()
```
