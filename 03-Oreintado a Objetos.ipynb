```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.9])
ax2 = fig.add_axes([0.2,0.55,0.4,0.3])

ax.plot(x,y,'b*--')
ax2.plot(y,x,'rs--')
ax2.set_facecolor('yellow')
fig.show()
```
