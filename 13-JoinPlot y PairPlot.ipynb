```python
import matplotlib.pyplot as plt
import seaborn as sns

tip = sns.load_dataset('tips')
tip.head(5)
```

```python
sns.jointplot(data=tip,x='total_bill',y='tip',hue='sex',kind='hist',
              marginal_ticks=True,marginal_kws=dict(bins=25,fill=True,multiple='dodge'))
plt.show()
```

```python
sns.pairplot(data=tip,hue='sex',palette='coolwarm',corner=True,kind='scatter')
plt.show()
```
