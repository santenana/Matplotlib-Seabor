```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tip = sns.load_dataset('tips')
tip.head(5)
```

```python
sns.histplot(data =tip,x='tip',bins=15,cumulative=False,
             hue='sex',stat='percent',multiple='dodge')
plt.show()
```

```python
sns.kdeplot(data=tip,x='tip',hue='sex',
            cumulative=False,fill=True,
            bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(data=tip,x='tip',hue='sex',stat='count',multiple='')
plt.show()
```

```python
sns.displot(data=tip,x='tip',hue='sex',kind='kde',fill=True)
plt.show()
```
