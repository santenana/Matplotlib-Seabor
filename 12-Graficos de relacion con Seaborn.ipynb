```python
import matplotlib.pyplot as plt
import seaborn as sns

tip = sns.load_dataset('tips')
tip.head(5)
```

```python
plt.figure(figsize=(8,8))
markers = {'Lunch':'D','Dinner':'s'}
sns.scatterplot(data=tip,x='total_bill',y='tip',hue='day',
                style='time',size='size',markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5 ))
plt.show()
```

```python
plt.figure(figsize=(8,8))
markers = {'Lunch':'*','Dinner':'^'}
sns.relplot(data=tip,x='total_bill',y='tip',hue='day',
            style='time',size='size',markers=markers,kind='scatter',col='sex')
plt.show()
```
