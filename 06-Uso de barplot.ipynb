```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
paises = ['India','Japon','Mexico','Colombia', 'Alemania']
poblacion = [1000,400,300,40,30]
```

```python
plt.bar(paises, poblacion,width=0.5,color=['#9932CC','#FFA07A','#4B0082','#6B8E23','#DAA520'])
plt.xticks(np.arange(5),('INDIA','JAPON','MEXICO','COLOMBIA','ALEMANIA'),rotation=45)
plt.title('Poblacion por paises')
plt.xlabel('Paises')
plt.ylabel('Poblacion en millones')
plt.show()
```

```python
plt.barh(paises, poblacion,color=['#9932CC','#FFA07A','#4B0082','#6B8E23','#DAA520'],height=0.5)
plt.yticks(np.arange(5),('INDIA','JAPON','MEXICO','COLOMBIA','ALEMANIA'),rotation=0)
plt.title('Poblacion por paises')
plt.ylabel('Paises')
plt.xlabel('Poblacion en millones')
plt.show()
```
