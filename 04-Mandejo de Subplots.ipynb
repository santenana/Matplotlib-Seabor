```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2,ncols=4)
ax[0,0].plot(x,y,'black')
ax[0,1].plot(x,np.cos(x),'red')
ax[0,2].plot(x,'yellow')
ax[0,3].plot(y,'bs-')
ax[1,0].plot(x,(x**2),'b*--')
ax[1,1].plot(x/2,'green')
ax[1,2].plot(x*y,'brown')
ax[1,3].plot(y,x,'red')
fig.tight_layout()
```
